```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import duckduckgo_search as duck
import requests
import time
from difflib import SequenceMatcher

def getHTMLdocument(url): 
    # request for HTML document of given url 
    response = requests.get(url) 
    # response will be provided in JSON format 
    return response.text 

def getImgUrl(name: str):
    browser = duck.DDGS()
    max_results=4
    
    try:
        answer = browser.text(name +' site:www.adrionltd.com',max_results=4 )
        url = (answer[0]).get('href')
   
    except Exception:
        answer = browser.text(name +' adrion ltd ',max_results=2 )
    
           
            

        

        
    index=0
    
    while (url[url.find('w'):url.find('m')+1]!= "www.adrionltd.com"):
        print("almost had me "+url)
        index=index+1
        if index>=max_results:
            url= "https://www.adrionltd.com"
            break
            
        else: url = (answer[index]).get('href')
        
            
        

        
    html_doc = getHTMLdocument(url)
    soup = bs(html_doc,'html.parser')
    imgUrl = soup.findAll(attrs={'class':'img-fluid'})
    # if not correct link out of range:
    imgUrl = (imgUrl[7]).get('src')
    return imgUrl
def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()
df = pd.read_csv(r"C:\Users\ThinkPad\ExtractedProjects\MjeksiaInventari1.csv", encoding='latin1')
df.columns=['title','author','year','price','SRC']
for index,row in df.iterrows():
        Title= row['title']
        Author= row['author']
        Title=Title.lower()
        time.sleep(1)
        try:
            SRC=getImgUrl(Title)
        except Exception:
            print("too fast")
        
        Title=Title.replace(" ","-")
        


        src_check=SRC
        tracker=src_check.find("default")+8
        src_check= src_check[tracker:-4]
        
        if(similar(Title,src_check)>0.59):
            print("FOUND: "+row['title']+"  "+SRC)
            
            row['SRC']= SRC

        
        else:
           print("FAIL "+Title)
           row['SRC']= "FAIL"

df.to_excel(r"C:\Users\ThinkPad\Documents\FinalURL.xlsx", index=False)
   
```
